```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path
import sqlite3
```

```python
# make db if it doesn't exist
Path('./data/group5.db').touch()
```

```python
# connect to db
conn = sqlite3.connect('./data/group5.db')
```

```python
# load the csv
base_df = pd.read_csv('./data/clean.csv')
```

```python
# write csv to db. if_exists will just drop the table if it already exists
base_df.to_sql(name='main',con=conn,if_exists='replace',index=False)
```
